```python
import numpy as np
import pandas as pd
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
filenames=[]
folder_path=r'c:\Users\jeetm\Desktop\ML - 100 Days\T20 Score Predictor\t20s'
for file in os.listdir(folder_path):
    filenames.append(os.path.join(folder_path,file))
```

```python
filenames[0:5]
```

```python
rows = []  # List of dicts (each row)

counter = 1
for file in tqdm(filenames):
    try:
        with open(file, 'r') as f:
            data = safe_load(f)
            if isinstance(data, dict):
                flat = pd.json_normalize(data)
                flat['match_id'] = counter
                rows.extend(flat.to_dict(orient='records'))  # ✅ Convert to row dicts
                counter += 1
    except:
        continue

# Now convert all rows at once to DataFrame
final_df = pd.DataFrame(rows)
```

```python
matches = pickle.load(open('dataset_level1.pkl','rb'))
matches.iloc[0]['innings'][0]['1st innings']['deliveries']
```

```python
count = 1
delivery_dfs = []  # Use a list to collect small dataframes

for index, row in matches.iterrows():
    if count in [75,108,150,180,268,360,443,458,584,748,982,1052,1111,1226,1345]:
        count += 1
        continue

    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []

    try:
        deliveries = row['innings'][0]['1st innings']['deliveries']
        batting = row['innings'][0]['1st innings']['team']
    except:
        count += 1
        continue  # Skip matches where structure is unexpected

    for ball in deliveries:
     for key in ball.keys():
        match_id.append(count)
        batting_team.append(batting)
        teams.append(row.get('info.teams'))
        ball_of_match.append(key)
        batsman.append(ball[key].get('batsman'))
        bowler.append(ball[key].get('bowler'))
        runs.append(ball[key].get('runs', {}).get('total', 0))
        city.append(row.get('info.city'))
        venue.append(row.get('info.venue'))

        # ✅ Fixed wicket block
        wicket_info = ball[key].get('wicket', {})
        if isinstance(wicket_info, dict):
            player_out = wicket_info.get('player_out', '0')
        else:
            player_out = '0'
        player_of_dismissed.append(player_out)



    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })

    delivery_dfs.append(loop_df)  # Collect the loop_df

    count += 1

# ✅ Final DataFrame
delivery_df = pd.concat(delivery_dfs, ignore_index=True)
```

```python
delivery_df
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
delivery_df['bowling_team'] = delivery_df.apply(bowl,axis=1)
```
